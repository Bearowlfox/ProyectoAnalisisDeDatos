```python
import pandas as pd
import numpy as np

df = pd.read_csv("DataSet.csv", low_memory=False, sep=";")

columns_to_normalize = ["Assessed Value", "Sale Amount", "Sales Ratio"]

df[columns_to_normalize] = df[columns_to_normalize].apply(pd.to_numeric, errors='coerce')

default_value = 0
df.fillna(default_value, inplace=True)

for col in columns_to_normalize:
    df[col + "_Normalized"] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

print(df.head())
```

```python
import pandas as pd

df = pd.read_csv("DataSet.csv", low_memory=False, sep=";")

bins = [0, 25, 50, 75, 100]
labels = ["Bajo", "Medio-Bajo", "Medio-Alto", "Alto"]

columns_to_bin = ["Assessed Value", "Sale Amount", "Sales Ratio"]

for col in columns_to_bin:
    df[col] = pd.to_numeric(df[col], errors='coerce')  

for col in columns_to_bin:
    df[col + "_Binning"] = pd.cut(df[col], bins=bins, labels=labels)

print(df.head())
```

<h5 Style = 'color:#8d3559'>  Combertir Valores Numericos en Funcion del Medio (Estandarizar)</h5>


```python
import pandas as pd

df = pd.read_csv("DataSet.csv", low_memory=False, sep=";")

columns_to_standardize = ["Assessed Value", "Sale Amount", "Sales Ratio"]

df_numeric = df[columns_to_standardize]

for col in df_numeric.columns:
    if pd.api.types.is_numeric_dtype(df_numeric[col]):
        mean = df_numeric[col].mean()
        std = df_numeric[col].std()

     
        df[col + "_Standardized"] = (df[col] - mean) / std

print(df.head())
```

```python
import pandas as pd

df = pd.read_csv("DataSet.csv", low_memory=False, sep=";")

bins = [0, 25, 50, 75, 100]
labels = ["Bajo", "Medio-Bajo", "Medio-Alto", "Alto"]

columns_to_bin = ["Assessed Value", "Sale Amount", "Sales Ratio"]

for col in columns_to_bin:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in columns_to_bin:
    df[col + "_Binning"] = pd.cut(df[col], bins=bins, labels=labels, right=True)

print(df.head())
```

```python
import pandas as pd

df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

df = df[df['Sales Ratio'].apply(pd.to_numeric, errors='coerce').notna()]

df['Sales Ratio'] = pd.to_numeric(df['Sales Ratio'], errors='coerce')

property_sales_ratio = df.groupby("Property Type")["Sales Ratio"].mean()

print(property_sales_ratio)

non_numeric_values = df[~df["Sales Ratio"].apply(lambda x: isinstance(x, (int, float)))]
print(non_numeric_values["Sales Ratio"].unique())

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

df = df[df['Sales Ratio'].apply(pd.to_numeric, errors='coerce').notna()]

df['Sales Ratio'] = pd.to_numeric(df['Sales Ratio'], errors='coerce')

property_sales_ratio = df.groupby("Property Type")["Sales Ratio"].mean()

top_property_sales_ratio = property_sales_ratio.nlargest(5)

plt.figure(figsize=(10, 6))
colors = plt.cm.Paired(range(len(top_property_sales_ratio)))
wedges, labels, autopct = plt.pie(top_property_sales_ratio, labels=top_property_sales_ratio.index, autopct=lambda pct: '{:.1f}%'.format(pct), colors=colors, startangle=140)
plt.title('Facilidad de Venta por Tipo de Propiedad - Pie Chart (Top 5)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.legend(wedges, ['{} ({:.1f}%)'.format(prop, pct) for prop, pct in zip(top_property_sales_ratio.index, top_property_sales_ratio)], title="Tipos de Propiedad", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(labels, fontsize=10)

plt.show()


```

<h2 Style = 'color:#63181f'>Pregunta 4. ¿Cuál es el costo promedio de las propiedades residenciales y 
comerciales?</h2>

```python
import pandas as pd

df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

residential_properties = df[df['Property Type'] == 'Residential']
commercial_properties = df[df['Property Type'] == 'Commercial']

# Verificar si la columna "Price" existe en el DataFrame
if 'Price' in residential_properties.columns and 'Price' in commercial_properties.columns:
    # Convertir los valores en la columna "Price" a numéricos solo si son válidos
    residential_properties['Price'] = pd.to_numeric(residential_properties['Price'], errors='coerce')
    commercial_properties['Price'] = pd.to_numeric(commercial_properties['Price'], errors='coerce')

    average_residential_cost = residential_properties['Price'].dropna().mean()
    average_commercial_cost = commercial_properties['Price'].dropna().mean()

    print(f"Costo promedio de propiedades residenciales: {average_residential_cost:.2f}")
    print(f"Costo promedio de propiedades comerciales: {average_commercial_cost:.2f}")
else:
    print("La columna 'Price' no existe en los datos filtrados.")

```

```python
import pandas as pd

df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

residential_properties = df[df['Property Type'] == 'Residential']
commercial_properties = df[df['Property Type'] == 'Commercial']

residential_properties['Sale Amount'] = pd.to_numeric(residential_properties['Sale Amount'], errors='coerce')
commercial_properties['Sale Amount'] = pd.to_numeric(commercial_properties['Sale Amount'], errors='coerce')

average_residential_cost = residential_properties['Sale Amount'].dropna().mean()
average_commercial_cost = commercial_properties['Sale Amount'].dropna().mean()

print(f"Costo promedio de propiedades residenciales: {average_residential_cost:.2f}")
print(f"Costo promedio de propiedades comerciales: {average_commercial_cost:.2f}")
```

<h2 Style = 'color:#63181f'>Pregunta 5. ¿Cuál es tipo de propiedades residenciales que se vende con mayor 
frecuencia, donde y con que costo promedio?</h2>

```python
import pandas as pd

df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

residential_properties = df[df['Property Type'] == 'Residential']

most_common_type = residential_properties['Town'].value_counts().idxmax()

most_common_properties = residential_properties[residential_properties['Town'] == most_common_type]

most_common_properties['Sale Amount'] = pd.to_numeric(most_common_properties['Sale Amount'], errors='coerce')

average_cost = most_common_properties['Sale Amount'].dropna().mean()

print(f"Tipo de propiedad residencial más frecuente: {most_common_type}")
print(f"Ubicación: {most_common_properties['Town'].iloc[0]}")
print(f"Costo promedio: {average_cost:.2f}")
```

<h2 Style = 'color:#63181f'>Pregunta 6.¿Cuál es margen promedio de ganancia por cada una de las

```python
import pandas as pd

df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

residential_properties_sold = df[df['Property Type'] == 'Residential']

most_common_type = residential_properties_sold['Town'].value_counts().idxmax()

most_common_properties = residential_properties_sold[residential_properties_sold['Town'] == most_common_type]

most_common_properties['Sale Amount'] = pd.to_numeric(most_common_properties['Sale Amount'], errors='coerce')

average_cost = most_common_properties['Sale Amount'].mean()

print(f"Tipo de propiedad residencial más frecuente: {most_common_type}")
print(f"Ubicación: {most_common_properties['Town'].iloc[0]}")
print(f"Costo promedio: {average_cost:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el DataFrame desde el archivo CSV
df = pd.read_csv('DataSet.csv', delimiter=';', low_memory=False)

# Filtrar propiedades residenciales vendidas
residential_properties_sold = df[df['Property Type'] == 'Residential']

most_common_type = residential_properties_sold['Town'].value_counts().idxmax()

most_common_properties = residential_properties_sold[residential_properties_sold['Town'] == most_common_type]

# Convertir la columna 'Sale Amount' a valores numéricos
most_common_properties['Sale Amount'] = pd.to_numeric(most_common_properties['Sale Amount'], errors='coerce')

average_cost = most_common_properties['Sale Amount'].mean()

# Crear un gráfico de barras
fig, ax = plt.subplots(figsize=(8, 6))
most_common_properties.plot(kind='bar', x='Property Type', y='Sale Amount', ax=ax, color='skyblue')

# Añadir título y etiquetas
plt.title('Venta de Propiedades Residenciales en Ubicación más Frecuente')
plt.xlabel('Tipo de Propiedad')
plt.ylabel('Monto de Venta')

# Mostrar el gráfico
plt.show()

# Imprimir los resultados
print(f"Tipo de propiedad residencial más frecuente: {most_common_type}")
print(f"Ubicación: {most_common_properties['Town'].iloc[0]}")
print(f"Costo promedio: {average_cost:.2f}")

```
